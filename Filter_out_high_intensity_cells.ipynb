```python
def drop_high_int (summary,df,max_int):
    df = df[df['intensity_mean'] > max_int].reset_index()
    print(df.shape)
    droprow = []
    for i in range(df.shape[0]):
        row = np.where((summary["roi"] == df["roi"][i]) & (summary['file name'].str.contains(df['file_name'][i])))[0].tolist()
        droprow=droprow+row
        #print(len(droprow))
    summary=summary.drop(droprow,axis=0).reset_index(drop=True)
    print('drop: ',len(droprow),';left: ', summary.shape[0])
    
    return summary
```

```python
def drop_low_traj (df,max_n):
    droprow = []
    for i in range(df.shape[0]):
        if (df.loc[i,'num_tracks_D'] < max_n):
            droprow.append(i)
    print(len(droprow))
    df=df.drop(droprow,axis=0).reset_index(drop=True)
    
    return df
```

```python
summary=drop_high_int (summary3,df,130)
alldata_f=drop_high_int (alldata,df,130)
```

```python
allangle_f=drop_high_int (allangle,df,130)
```
