```python
import numpy as np
from os import listdir, path, system
import cv2
import matplotlib.pyplot as plt
import math
import matplotlib.patches as patches
import os
import csv
import matplotlib.image as mpimg
from skimage import io
import imageio
#conda install ...
```

```python
def find_background(data_dir,file):
    img1 = io.imread(f'{data_dir}/{file}')
    #./0421P1CM new OE/scan_Top Slide_D_p00_0_A01f00d0.TIF
    row,col=img1.shape # (128,256)
    #num=np.zeros(256)
    s=np.zeros((row,col))
    k=25
    maxnum=65535*k*k #1e180
    maxx=-1
    maxy=-1
    #print(row,col)
    for y in range(0,row):
        for x in range(0,col):
            #img1(x,y) 小框框(pixel)
            # s(x,y) 大框框( (0,0) -> (x,y) )
            if y*x != 0:
                s[y][x]=s[y-1][x]+s[y][x-1]-s[y-1][x-1]+img1[y][x]
            elif x==0 and y!=0:
                s[y][x]=img1[y][x]+s[y-1][x]
            elif y==0 and x!=0:
                s[y][x]=img1[y][x]+s[y][x-1]
            else:
                s[y][x]=img1[y][x]
            #中框框(k*k)
            if x>=k and y>=k:
                test = s[y][x]-s[y-k][x]-s[y][x-k]+s[y-k][x-k]
                if test < maxnum:
                    maxx=x
                    maxy=y
                    maxnum=test
                    #print(x,y,maxnum)
    print(maxx,maxy,maxnum/k/k)
    plt.imshow(img1,'gray')
    currentAxis=plt.gca()
    rect=patches.Rectangle((maxx-k-1,maxy-k-1), k+2, k+2,linewidth=1,fill=False, edgecolor = '#FFFFFF')
    currentAxis.add_patch(rect)
    plt.axis('off') 
    plt.savefig(data_dir+'/background/'+file[0:-4]+'.png',dpi=540,bbox_inches = 'tight')
    #plt.show()
    plt.clf()
    data.append([file, maxnum/k/k, maxnum, maxx, maxy]) #follow the order of imagej result.csv
```

```python
'''folder="0421P1CM new OE" #資料夾名稱
data_dir="./"+folder #相對路徑'''
data_dir='/Volumes/Holtlab_YST/Yi-Shuan/microscope/2022/2022_Oct-Dec/20221128_hPNE_5EU_LMB/5EU' #絕對路徑
all_file=listdir(data_dir)
data=[["Label","Mean","RawInt", "x", "y"]]
for file in all_file:  #need to modify for different files
    if '.tif' in file:
        print(file)
        find_background(data_dir,file)
with open(data_dir+'/background/BGInt_Results.csv', 'w',newline ='') as f:  
    writer = csv.writer(f) 
    writer.writerows(data) 
    #csv.writer(f).writerows(data)
```
